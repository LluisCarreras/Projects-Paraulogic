```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.keys import Keys
import time
```

```python
url = "https://dlc.iec.cat/"
```

```python
# Set up the WebDriver
driver = webdriver.Chrome()  # Use the appropriate driver for your browser
driver.get(url)  # Replace with the target web page URL
```

```python
alphabet = 'abcçdefghijklmnopqrstuvwxyz'

# List with all the possible bigraphs that can be made with the catalan alphabet
bichars = []

for first_char in alphabet:
    for second_char in alphabet:
        new_bichar = f'{first_char}{second_char}'
        bichars.append(new_bichar)

len(bichars)
```

```python
words = []
```

```python
def get_words(chars):

    try:
        # Set up the WebDriver
        driver = webdriver.Chrome()  # Use the appropriate driver for your browser
        driver.get(url)  

        # Interact with the Text Input Field
        text_input = driver.find_element(By.XPATH, '//input[@type="text"]') 
        text_input.send_keys(chars) 

        # Interact with the Select Field
        select_element = driver.find_element(By.XPATH, '//select')  
        select = Select(select_element)
        select.select_by_visible_text("Començada per") 

        # Click the Submit Field
        submit_button = driver.find_element(By.XPATH, '//input[@type="submit"]')  
        submit_button.click() 

        # Wait for the Results to Load
        time.sleep(5)  

        # Scrape the Resulting List
        result_elements = driver.find_elements(By.XPATH, '//div[@class="resultText"]')  # Adjust XPath
        results = [element.text for element in result_elements]

        # Close the WebDriver
        driver.quit()

        return results
    except:
        return []

    
```

```python
def get_trichars(bichar):
    trichars = [f'{bichar}{char}' for char in alphabet]
    return trichars
```

```python
for bichar in bichars:

    new_words = get_words(bichar)
    print(f'{bichar}: {len(new_words)}')
    
    # Add the results to the words list
    if len(new_words) < 1000:
        words.extend(new_words)
    else:
        trichars = get_trichars(bichar)
        for trichar in trichars:
            new_words = get_words(trichar)
            print(f'\t{trichar}: {len(new_words)}')
            words.extend(new_words)
```

```python
import csv

file_path = 'raw_words.csv'

# Save to CSV
with open(file_path, "w", encoding="utf-8", newline="") as csvfile:
    writer = csv.writer(csvfile)
    for word in words:
        writer.writerow([word])  # Each item in a new row

print("CSV file created with the new data.")
```
